# Array insertion and deletion

```python
import array
import numpy as np

array1 = array.array('i', [971,972])
array2 = array.array('i', [1,2,3,5,6,7,9])
#print(np_array2[-1])

#another way array can be defined using other liberary
array3 = np.array([1,5,7,9,9,10,'ab'])

#array insertion
array2.insert(0, 100)
array2.insert(2, 45)
#deleting element from array
array2.remove(45)
array2
```

# Linear search

```python
#linear search
def linearsearch(are, targt):
    for i in range(len(are)):
        if are[i] == targt:
            return i
    return -1
print(linearsearch(array2, 7))
```

# Array Traverse

```python
#array traversal
def traverse(ar):
    are = ar.tolist()
    for i in are:
        print(i)

traverse(array2)
```

#Multi dimentional array

```python
# Day 1 - 11, 15, 9, 6
# Day 2 - 15, 24, 20, 26

np_array = np.array([[11,15,9,6],[15,24,20,26]])
multi_array = np.insert(np_array, 0, [[1,2, 3, 4]], axis=0)
print(multi_array)

#Access value in multidimentional array

def accessElements(array, rowIndex, colIndex):
  if rowIndex >= len(array) and colIndex >= len(array[0]):
    print('Incorrect Index')
  else:
    print(array[rowIndex][colIndex])

accessElements(multi_array, 1, 2)
```

#Traversing multidimentional array

```python
def traverseArray(array):
  for i in range(len(array)):
    for j in range(len(array[i])):
      print(array[i][j])

traverseArray(multi_array)
```

#Searching in multidimentional array

```python
def multi_linear_search(array, value):
  for i in range(len(array)):
    for j in range(len(array[i])):
      if array[i][j] == value:
        return 'The value is located at index ' + str(i) + ' ' + str(j)
  return 'The element is not found'

print(multi_linear_search(multi_array, 24))
```

#Deleting row/column from multi dimentional array

```python
print(multi_array)
new_array = np.delete(multi_array, 0, axis=0)
print(new_array)
```
